# paquete numpy

```python
import numpy as np
import numpy.linalg as ln
```

## np.array

Con esto se pueden crear vectores, matrices y tensores.

```python
vector=np.array([1,2,3,4])

matriz=np.array([[1,2],
                 [3,4]])

print(f"Vector:\n {vector}")
print("-"*100)
print(f"Matriz:\n {matriz}")
```

## $\text{np.zeros}(t)$

### ¿Que hace ?

Para crear el vector nulo o la matriz nula.

### Parametros

$\bullet$ $t$ es la tupla que define las filas y las colummnas del array.


```python
vector_nulo=np.zeros(3)
matriz_nula=np.zeros((3,2))

print(f"Vector nulo:\n {vector_nulo}")
print("-"*100)
print(f"Matriz nula:\n {matriz_nula}")
```

## $\text{np.ones}(t)$

### ¿ Que hace ?

Para crear un vector o una matriz solo de unos.

### Parametros

$\bullet$ $t$ es la tupla que define las filas y las colummnas del array.

```python
vector=np.ones(3)
matriz=np.ones((3,2))

print(f"Vector de unos:\n {vector}")
print("-"*100)
print(f"Matriz de unos:\n {matriz}")
```

## $\text{np.arange}(a,b,c)$

### ¿ Que hace ?

Crea una secuencia en forma vector.

### Parametros

$\bullet$ $a$ y $b$ forman el valor inicial y final de la secuencia.

$\bullet$ $c$ es el número de terminos.

```python
secuencia=np.arange(0,12,2)
print(f"Secuencia:\n {secuencia}")
```

## $\text{np.linespace}(a,b,c)$

### ¿ Que hace ?

Elejir términos de un intervalo.

### Parametros

$\bullet$ $a$ y $b$ forman el intervalo $[a,b]$ de donde se tomaran los términos.

$\bullet$ $c$ es el número de terminos.

```python
puntos=np.linspace(-2,2,5)
print(f"Puntos:\n {puntos}")
```

## $\text{np.logspace}(a_1,a_n,c)$

### ¿ Que hace ?

Crear una secuencia del tipo $10^n$.

### Parametros:

$\bullet$ $a_1$ y $a_n$ forman el conjunto $\{a_1 < a_2<\dots < a_n\}$ de exponentes donde $a_i\in\mathbb{N}$

$\bullet$ $c$ es el número de términos de la forma $10^n$.

```python
secuencia=np.logspace(0,10,5)
print(f"Secuencia logaritmica:\n {secuencia}")
```

## $\text{np.meshgrid}(x,y)$

### ¿ Que hace ?

Crea dos matrices a partir de dos vectores, la cual sirve para evaluar funciones de dos variables y crear graficas 3D.

### Parametros:

$\bullet$ $x,y$ son los array.

```python
x=np.array([1,2,12,-2])
y=np.array([1,-6,8,0])

X,Y=np.meshgrid(x,y)

print(f"X:\n {X}")
print("-"*100)
print(f"Y:\n {Y}")
print("-"*100)

# Donde Z[i,j] = X[i,j]**2 + Y[i,j]**2
Z=X**2+Y**2
print(f"Z:\n {Z}")
```

## $\text{np.fromfunction}(\text{ \text{lambda} , (n,m)})$

Para crear una matriz donde cada entrada corresponde a un función $f(i,j)$, donde $i,j$ es la fila y columna de la matriz.

Se define la función y luego se da el tamaño de la matriz.

```python
matriz=np.fromfunction(lambda i,j: i*j, (4,5))

print(f"Matriz:\n {matriz}")
```

## $\text{np.reshape}(a,b)$

### ¿ Que hace?

Modificar el numero de filas o columnas de vector y una matriz.

### Parametros:

$\bullet$ $a$ y $b$ son el número de filas y columnas que se desean.


```python
matriz=np.array([[1,2,3],
                 [4,5,6]])

matriz_a_dos_columnas=matriz.reshape(3,2)
matriz_a_una_fila=matriz.reshape(6)

print(f"Matriz a dos columnas:\n {matriz_a_dos_columnas}")
print("-"*100)
print(f"Matriz a una fila:\n {matriz_a_una_fila}")
```

## $a\text{.flatten}$

### ¿ Que hace ?

Convertir una matriz en un vector.

### Parametros:

$\bullet$ $a$ es el array.

```python
matriz=np.array([[1,2],
                 [3,4]])
matriz_a_vector=matriz.flatten()
print(f"Matriz a vector:\n {matriz_a_vector}")
```

## np.vstack

Usar vectores como filas de una matriz.

Unir matrices en una sola

```python
fila_1=np.array([1,2,3])

fila_2=np.array([-4,-5,-6])

matriz_1=np.array([[-1,-4,-5],
                   [1,2,3]])

matriz_2=np.array([[11,27,-3],
                   [4,-5,6]])

matriz_con_vectores=np.vstack((fila_1,fila_2))

Matriz_con_matrices=np.vstack((matriz_1,matriz_2))

print(f"Matriz con vectores:\n {matriz_con_vectores}")
print("-"*100)
print(f"Matriz con matrices:\n {Matriz_con_matrices}")
```

## np.hstack

Unir vectores en uno solo

Unir matrices en una sola

```python
vector_1=np.array([1,2,3])

vector_2=np.array([-4,-5,-6])

matriz_1=np.array([[-1,-4,-5],
                   [1,2,3]])

matriz_2=np.array([[11,27,-3],
                   [4,-5,6]])

matriz_con_vectores=np.hstack((vector_1,vector_2))

Matriz_con_matrices=np.hstack((matriz_1,matriz_2))

print(f"Matriz con vectores:\n {matriz_con_vectores}")
print("-"*100)
print(f"Matriz con matrices:\n {Matriz_con_matrices}")
```

## np.append

Agregar elementos a vectores o matrices.

La primera entrada es el array, el segundo es el elemento a agregar y el tercero es la forma de agregar el elemento, donde solo se puede poner $\text{axis}=0$ o $\text{axis}=1$.

```python
vector=np.array([1,2,3])
matriz=np.array([[1,2,3],
                [4,5,6]])

data1=np.array([15,25,35,45])
data2=np.array([[-1,-2,-3],
                [14,15,16]])

vector_data1=np.append(vector,data1)
matriz_data2=np.append(matriz,data2,axis=1)

print(f"Vector con data1:\n {vector_data1}")
print("-"*100)
print(f"Matriz con data2:\n {matriz_data2}")
```

## Operaciones de matrices

Solo las entrada por entrada

```python
matriz1=np.array([[-1,2,3],
                 [4,-5,0]])
matriz2=np.array([[1,8,-9],
                 [10,-11,12]])

matriz_suma=matriz1+matriz2
matriz_resta=matriz1-matriz2
matriz_producto=matriz1*matriz2 
matriz_division=matriz1/matriz2
Matriz_potencia=matriz1**2

print(f"Matriz suma:\n {matriz_suma}")
print("-"*100)
print(f"Matriz resta:\n {matriz_resta}")
print("-"*100)
print(f"Matriz producto:\n {matriz_producto}")
print("-"*100)
print(f"Matriz division:\n {matriz_division}")
print("-"*100)
print(f"Matriz potencia:\n {Matriz_potencia}")
```
